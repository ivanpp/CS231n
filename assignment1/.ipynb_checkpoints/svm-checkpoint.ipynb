## CIFAR-10 Data Loading and Preprocessing

````python
# Load the raw CIFAR-10 data.
cifar10_dir = 'cs231n/datasets/cifar-10-batches-py'
X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)

# As a sanity check, we print out the size of the training and test data.
print('Training data shape: ', X_train.shape)
print('Training labels shape: ', y_train.shape)
print('Test data shape: ', X_test.shape)
print('Test labels shape: ', y_test.shape)
````

````python
# Visualize some examples from the dataset.
# We show a few examples of training images from each class.
classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
num_classes = len(classes)
samples_per_class = 7
for y, cls in enumerate(classes):
    idxs = np.flatnonzero(y_train == y)
    idxs = np.random.choice(idxs, samples_per_class, replace=False)
    for i, idx in enumerate(idxs):
        plt_idx = i * num_classes + y + 1
        plt.subplot(samples_per_class, num_classes, plt_idx)
        plt.imshow(X_train[idx].astype('uint8'))
        plt.axis('off')
        if i == 0:
            plt.title(cls)
plt.show()
````

````python
# Split the data into train, val, and test sets. In addition we will
# create a small development set as a subset of the training data;
# we can use this for development so our code runs faster.
num_training = 49000
num_validation = 1000
num_test = 1000
num_dev = 500

# Our validation set will be num_validation points from the original
# training set.
mask = range(num_training, num_training + num_validation)
X_val = X_train[mask]
y_val = y_train[mask]

# Our training set will be the first num_train points from the original
# training set.
mask = range(num_training)
X_train = X_train[mask]
y_train = y_train[mask]

# We will also make a development set, which is a small subset of
# the training set.
mask = np.random.choice(num_training, num_dev, replace=False)
X_dev = X_train[mask]
y_dev = y_train[mask]

# We use the first num_test points of the original test set as our
# test set.
mask = range(num_test)
X_test = X_test[mask]
y_test = y_test[mask]

print('Train data shape: ', X_train.shape)
print('Train labels shape: ', y_train.shape)
print('Validation data shape: ', X_val.shape)
print('Validation labels shape: ', y_val.shape)
print('Test data shape: ', X_test.shape)
print('Test labels shape: ', y_test.shape)
````

````python
# Preprocessing: reshape the image data into rows
X_train = np.reshape(X_train, (X_train.shape[0], -1))
X_val = np.reshape(X_val, (X_val.shape[0], -1))
X_test = np.reshape(X_test, (X_test.shape[0], -1))
X_dev = np.reshape(X_dev, (X_dev.shape[0], -1))

# As a sanity check, print out the shapes of the data
print('Training data shape: ', X_train.shape)
print('Validation data shape: ', X_val.shape)
print('Test data shape: ', X_test.shape)
print('dev data shape: ', X_dev.shape)
````

````python
# Preprocessing: subtract the mean image
# first: compute the image mean based on the training data
mean_image = np.mean(X_train, axis=0)
print(mean_image[:10]) # print a few of the elements
plt.figure(figsize=(4,4))
plt.imshow(mean_image.reshape((32,32,3)).astype('uint8')) # visualize the mean image
plt.show()
````

````python
# second: subtract the mean image from train and test data
X_train - mean_image
X_val - mean_image
X_test - mean_image
X_dev -= mean_image
````

````python
# third: append the bias dimension of ones (i.e. bias trick) so that our SVM
# only has to worry about optimizing a single weight matrix W.
X_train = np.hstack([X_train, np.ones((X_train.shape[0], 1))])
X_val = np.hstack([X_val, np.ones((X_val.shape[0], 1))])
X_test = np.hstack([X_test, np.ones((X_test.shape[0], 1))])
X_dev = np.hstack([X_dev, np.ones((X_dev.shape[0], 1))])

print(X_train.shape, X_val.shape, X_test.shape, X_dev.shape)
````

It is possible that once in a while a dimension in the gradcheck will not match exactly. What could such a discrepancy be caused by? Is it a reason for concern? **Your Answer:** 
1. This question caused by the surge of the 'maximum function'. If we increase the value of one single scalar w in the matrix W for just a little bit, some elements in the output matrix(score matrix) will increase/decrease a liitle bit(After we subtract the mean image from the dataset, the input matrix X is no longer non-negative). The little increase will probably 'active(or extinct)' the maximum function. As a result, the numerical grad will increase(or decrease) accordingly. And that increase/decrease will be very tiny. But for the analytic grad, the situation will be differnt. The relevant input vector x will be added to some specific column. And the case of the ground-truth column will be the opposite. For that specific dw we compared with the numerical grad, it will be added or subtracted from a larger value(from '-mean image' to '255-mean image')

![image2 for inline quesetion by ivanpp](http://oo1ncxa8y.bkt.clouddn.com//markdown-img-paste-20171010215212377.png)
That is beacuse for these similar number it is more likely to active(or extinct) the maximum fuction. And numerous activations will collapse the numerical grad while the analytic grad may function well. This is because too many activations interact with each other. Add one x[i], subtract one x[i], add one x[i]...    
The offset helps the analytic grad maintain its well-function in this case.
3. If it is case 1. I think it's no big deal. The numerical grad and analytic grad may be different but the difference is mostly marginal. And the direction will be right. Some w in W matrix may be improved a bit larger than it should be. But in the next iteration, this runoff will be fixed(if it is still need to be fixed).    
If it is case 2. This is due to the poor initialization. We should initialize the W matrix with some more appropriate value. At least we can save some time in the training progress.
4. Here is a clear example:    
Suppose that it is w from row 2 and col 3 that we compute. 'w' increases a little bit, then the col3(represent class3) of the  output matrix(score) will be changed. Then the negligible change just makes something happend. One certain margin increases from -0.125 to 0.007.

````python
# Complete the implementation of svm_loss_vectorized, and compute the gradient
# of the loss function in a vectorized way.

# The naive implementation and the vectorized implementation should match, but
# the vectorized version should still be much faster.
tic = time.time()
_, grad_naive = svm_loss_naive(W, X_dev, y_dev, 0.000005)
toc = time.time()
print('Naive loss and gradient: computed in %fs' % (toc - tic))

tic = time.time()
_, grad_vectorized = svm_loss_vectorized(W, X_dev, y_dev, 0.000005)
toc = time.time()
print('Vectorized loss and gradient: computed in %fs' % (toc - tic))

# The loss is a single number, so it is easy to compare the values computed
# by the two implementations. The gradient on the other hand is a matrix, so
# we use the Frobenius norm to compare them.
difference = np.linalg.norm(grad_naive - grad_vectorized, ord='fro')
print('difference: %f' % difference)
````

### Inline question 2:
Describe what your visualized SVM weights look like, and offer a brief explanation for why they look they way that they do.

**Your answer:**     
The visualized SVM weights are very abstract. The 'plane' looks sort of symmetric. The 'frog' looks like some smashed green things. The 'horse' has a red upper half and green for the bottom half.    
f = Wx, W has the shape of (dim, num_class). And W is optimized to have the value to let the 'horse' sample to get num_class scores for num_class classes, highest for 'horse' certainly.    
So for one specific slice of the W matrix with the shape of (dim, 1), if i feed it with one sample(1, dim) from that specific class. It will give a  score as high as possible. Since the input images have been normolized(subtract the image mean), they are no longer 'all positive'. So the simplest way to have the highest score is to be similar to the input(at least for the signs).(e.g. 0.05\*0.05+(-0.4)\*(-0.4))
So the final W will be similar to the average of the image samples subtract the image mean. But they will never be the average of the specific class samples subtract the image meam because they interact with each other.    
So the 'horse' weight is 40% of horses add 60% of others then subtract the image mean.(Suppose accuracies for all classes are 0.4)
